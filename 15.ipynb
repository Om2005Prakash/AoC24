```python
with open("in15.txt", "r") as f:
    #Preparing Map and Moves Input
    data = f.readlines()
    map = []
    moves = []
    for row in data:
        if row[0] == '#':
            temp = []
            for j in row[:-1]:
                if j == "#":
                    temp.extend(["#","#"])
                elif j == "@":
                    temp.extend(["@","."])
                elif j == ".":
                    temp.extend([".","."])
                elif j == "O":
                    temp.extend(["[","]"])
            map.append(temp)
        elif row[0] in '<^>v':
            moves.extend(list(row[:-1]))

    dirs = [(-1,0),(0,1),(1,0),(0,-1)]
    map_dir = {"^": (-1,0), ">":(0,1), "v":(1,0), "<":(0,-1)}
    width = len(map[0])
    height = len(map)

    def check(j, i, dir):
        global nodes
        if map[j][i] == "#":
            return 0
        elif map[j][i] == ".":
            return 1
        nj = j + dir[0]
        ni = i + dir[1]
        if dir[1] == 0:
            if map[j][i] == "[" and map[j][i+1] == "]" and check(nj, ni, dir) and check(nj, ni + 1, dir):
                nodes.add((j,i))
                nodes.add((j,i+1))
                return 1
            elif map[j][i] == "]" and map[j][i-1] == "[" and check(nj, ni, dir) and check(nj, ni-1, dir):
                nodes.add((j,i))
                nodes.add((j,i-1))
                return 1
            else:
                return 0
        else:
            if check(nj, ni, dir):
                nodes.add((j,i))
                return 1
            else:
                return 0

    curr = [0, 0]

    for j in range(height):
        for i in range(width):
            if map[j][i] == "@":
                curr = [j, i]
                map[j][i] = "."

    for move in moves:
        dir = map_dir[move]
        nj = curr[0] + dir[0]
        ni = curr[1] + dir[1]
        if map[nj][ni] == ".":
            curr = [nj, ni]
        else:
            nodes = set()
            if check(nj, ni, dir):
                curr = [nj, ni]
                map_temp = [[map[j][i] if (j,i) not in nodes else '.' for i in range(width)] for j in range(height)]
                for node in nodes:
                    j = node[0]
                    i = node[1]
                    nj = j + dir[0]
                    ni = i + dir[1]
                    map_temp[nj][ni] = map[j][i]
                map = [[map_temp[j][i] for i in range(width)] for j in range(height)]
    res = 0

    for j in range(height):
        for i in range(width):
            if map[j][i] == "[" and map[j][i+1] == "]":
                res+=(j)*100 + (i)
    print("Part 2", res)
```
