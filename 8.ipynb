```python
with open("in8.txt", "r") as f:
    data = f.readlines()
    data = [list(i[:-1]) for i in data]
    data_cpy = [i.copy() for i in data]
    antenas = dict()
    length = len(data[0])
    width = len(data)

    for j in range(width):
        for i in range(length):
            if data[j][i] != '.':
                if data[j][i] in antenas:
                    antenas[data[j][i]].append((j,i))
                else:
                    antenas[data[j][i]] = [(j,i)]
    
    potential = []
    for antena in antenas:
        group = antenas[antena]
        for j in group:
            for i in group:
                if j!=i:
                    k = 1
                    if (0<=(k*(i[0] - j[0]) + i[0])<width and 0<=(k*(i[1] - j[1]) + i[1])<length):
                        a = (k*(i[0] - j[0]) + i[0], k*(i[1] - j[1]) + i[1])
                        potential.append((antena, a))
    res = 0
    for s,i in potential:
        if 0<=i[0]<width and 0<=i[1]<length:
            if (data_cpy[i[0]][i[1]] != "#"):
                res += 1
                data_cpy[i[0]][i[1]] = "#"
    
print("Part 1:",res) 
```

```python
for i in data_cpy:
    for j in i:
        print(j,end='')
    print()
```
