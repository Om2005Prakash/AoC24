```python
import sys
sys.setrecursionlimit(int(1e8))

with open("in20.txt", "r") as f:
    track = f.readlines()
    track = [list(i[:-1]) for i in track]

    width = len(track[0])
    height = len(track)

    for j in range(height):
        for i in range(width):
            if track[j][i] == "S":
                start = (j,i)
                track[j][i] = "."
            elif track[j][i] == "E":
                end = (j,i)
                track[j][i] = "."

    dirs = [(-1,0), (0,1), (1,0), (0,-1)]
    track_bool = [[0 for _ in range(width)] for _ in range(height)]
    cheats = [(j, i) for j in range(-20,21) for i in range(-20,21) if abs(i) + abs(j) < 21]
    cache = [[float('inf') for _ in range(width)] for _ in range(height)]

    sols = []
    track_bool[start[0]][start[1]] = 1

threshold = 9332

def dfs(j, i, n, cheat):
    global track_bool
    global sols

    if n > threshold:
        return

    if (cheat == 1) and (cache[j][i] != -1) and (n + cache[j][i] < threshold):
        sols.append(n + cache[j][i])
        return

    if (j, i) == end:
        cache[j][i] = 0
        sols.append(n)
        return
    
    temp = float('inf')

    for dir in dirs:
        nj = j + dir[0]
        ni = i + dir[1]
        possible = n + abs(end[0] - nj) + abs(end[1] - ni) - threshold
        if (possible <= 0) and (0<=nj<height) and (0<=ni<width) and (track_bool[nj][ni] == 0) and (track[nj][ni] == '.'):
                track_bool[nj][ni] = 1
                dfs(nj, ni, n+1, cheat)
                track_bool[nj][ni] = 0
                if cache[nj][ni] != -1:
                    temp = min(temp, cache[nj][ni])

    if cheat:
        cache[j][i] = 1 + temp

    if cheat == 0:
        for (dj, di) in cheats:
            nnj = j + dj
            nni = i + di
            possible = n + abs(end[0] - nnj) + abs(end[1] - nni) - threshold
            if (possible <= 0) and 0<=nnj<height and 0<=nni<width and (track[nnj][nni] == '.') and (track_bool[nnj][nni] == 0):
                track_bool[nnj][nni] = n+abs(dj)+abs(di)
                dfs(nnj, nni, n+abs(dj)+abs(di), 1)
                track_bool[nnj][nni] = 0

dfs(start[0], start[1], 0, 0)
print(len(sols))
```
