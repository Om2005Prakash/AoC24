```python
# import re

# text = "There are 123 apples"

# all_matches = re.findall(r"\w+", text)  # Matches all words
# print(all_matches)  # ['There', 'are', '123', 'apples']


with open("in2.txt", "r") as f:
    data = f.readlines()
    res = 0

    for i in data:
        levels = [int(j) for j in i.split()]
        flag_inc = 0
        flag_dec = 0

        j = 0
        while j < len(levels) - 1:
            if 1<= levels[j+1] - levels[j] <= 3:
                j = j + 1
            else:
                if j > 1 and j + 2 < len(levels):
                    temp_1 = levels[j+2] - levels[j]
                    temp_2 = levels[j+1] - levels[j-1]
                    if temp_1 < 1 or temp_1 > 3:
                        if temp_2 < 1 or temp_2 > 3:
                            flag_inc += 100
                            break
                        j = j + 1
                        continue
                    else:
                        j = j +2
                        continue
                elif j == 0:
                    temp_1 = levels[j+2] - levels[j]
                    if temp_1 < 1 or temp_1 > 3:
                        flag_inc += 100
                        break
                    j = j + 2
                    continue
                else:
                    temp_2 = levels[j+1] - levels[j-1]
                    if temp_2 < 1 or temp_2 > 3:
                        flag_inc+=100
                        break
                    j = j + 1
        
        levels.reverse()
        j = 0
        while j < len(levels) - 1:
            if 1<= levels[j+1] - levels[j] <= 3:
                j = j + 1
            else:
                if j > 1 and j + 2 < len(levels):
                    temp_1 = levels[j+2] - levels[j]
                    temp_2 = levels[j+1] - levels[j-1]
                    if temp_1 < 1 or temp_1 > 3:
                        if temp_2 < 1 or temp_2 > 3:
                            flag_dec += 100
                            break
                        j = j + 1
                        continue
                    else:
                        j = j +2
                        continue
                elif j == 0:
                    temp_1 = levels[j+2] - levels[j]
                    if temp_1 < 1 or temp_1 > 3:
                        flag_dec += 100
                        break
                    j = j + 2
                    continue
                else:
                    temp_2 = levels[j+1] - levels[j-1]
                    if temp_2 < 1 or temp_2 > 3:
                        flag_dec+=100
                        break
                    j = j + 1
        
        flag = min(flag_inc, flag_dec)
        # print(flag_inc, flag_dec)
        if flag < 2:
            res += 1
    print(res)
```
