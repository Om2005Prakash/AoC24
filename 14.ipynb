```python
import re
from collections import deque

connected = 0

def bfs(j, i, condition):
        queue = deque([(j, i),])
        count = 0
        map_bool[j][i] = 1

        while queue:
            (j, i) = queue.popleft()
            for dir in dirs:
                nj = j + dir[0]
                ni = i + dir[1]
                if (0<=nj<tall) and (0<=ni<wide) and condition(nj, ni, j, i) and map_bool[nj][ni] == 0:
                    count+=1
                    queue.append((nj, ni))
                    map_bool[nj][ni] = 1
        return count

def test(nj, ni, j, i):
    return map[nj][ni] > 0

with open("in14.txt", "r") as f:
    wide = 101
    tall = 103
    pos = f.readlines()
    pos = [re.findall(r'[+-]?\d+', i) for i in pos]
    pos = [[int(j) for j in i] for i in pos]
    t = 0
    dirs = [(-1,0),(1,0),(0,1),(0,-1)]
    while True:
        t+=1
        map = [[0 for _ in range(wide)] for _ in range(tall)]
        map_bool = [[0 for _ in range(wide)] for _ in range(tall)]
        for robot in pos:
            j = (robot[1] + t*robot[3])%tall
            i = (robot[0] + t*robot[2])%wide
            map[j][i]+=1

        for j in range(tall):
            for i in range(wide):
                if map[j][i] > 0 and map_bool[j][i]!=1:
                    res = bfs(j, i, test)
                    if res > connected:
                        connected = res
                        print(t, connected)
        
```
