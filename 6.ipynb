```python
with open("in6.txt" , "r") as f:

    '''
    For Part 1,
    We simply simulate the behaviour
    '''

    map = f.readlines()
    map = [list(i[:-1]) for i in map]
    seen = [[-1 for i in range(len(map[0]))] for j in range(len(map))]

    curr = [0,0]
    vectors = [[-1,0],[0,1],[1,0],[0,-1]]
    ind = 0

    for j in range(len(map)):
        for i in range(len(map[0])):
            if map[j][i] == '^':
                curr = [j,i]
                ind = 0
                map[j][i] = '.'
            elif map[j][i] == '>':
                curr = [j,i]
                ind = 1
                map[j][i] = '.'
            elif map[j][i] == 'v':
                curr = [j,i]
                ind = 2
                map[j][i] = '.'
            elif map[j][i] == '<':
                curr = [j,i]
                ind = 3
                map[j][i] = '.'
    
    seen[curr[0]][curr[1]] = ind

    count = 0
    nj = curr[0] + vectors[ind][0]
    ni = curr[1] + vectors[ind][1]
    while (0<=nj<len(map)) and (0<=ni<len(map[0])):
        if map[nj][ni] == '.':
            curr = [nj, ni]
            if seen[nj][ni] == -1:
                seen[nj][ni] = ind
                count+=1
        else:
            ind = (ind + 1)%4
        nj = curr[0] + vectors[ind][0]
        ni = curr[1] + vectors[ind][1]

    print("Part 1", count)

    '''
    For Part 2,
    We should try to place obstacle at each nodes accessable by the guard,
    and try to run guard and see if it gets stuck in a loop
    for checking loop, we can make a new temp_seen array which is -1 everywhere and it is updated as guard move
    If a node has already been seen then we might come it to "loop". To check this we see if the previsious visit
    and current visit have same direction, if yes then we are stuck

    (NOTE, in my implementation seen array contains for each node, direction to move from its parent node)
    (i.e we can trace the parent for each node)
    '''

    res = 0
    for j in range(len(map)):
        for i in range(len(map[0])):
            if seen[j][i] != -1:
                ind = seen[j][i]
                nj = j
                ni = i
                curr = [j - vectors[ind][0], i - vectors[ind][1]]
                temp_seen = [[-1 for i in range(len(map[0]))] for j in range(len(map))]
                map_cpy = [[elem for elem in row] for row in map]                       #deep copy
                map_cpy[j][i] = '#'

                flag = 0
                while (0<=nj<len(map)) and (0<=ni<len(map[0])):
                    if map_cpy[nj][ni] == '.':
                        curr = [nj, ni]
                        if temp_seen[nj][ni] == -1:
                            temp_seen[nj][ni] = ind
                        else:
                            if temp_seen[nj][ni] == ind:
                                flag = 1
                                break
                    else:
                        ind = (ind + 1)%4
                    nj = curr[0] + vectors[ind][0]
                    ni = curr[1] + vectors[ind][1]

                if flag:
                    res += 1

    print("Part 2", res)
```
