```python
import heapq

with open("in16.txt", "r") as f:
    map = f.readlines()
    map = [list(i[:-1]) for i in map]
    map_cost = [[10_000 for i in j] for j in map]
    length = len(map[0])
    height = len(map)
    
    dirs = [(-1,0), (0,1), (1,0), (0,-1)]

    for j in range(height):
        for i in range(length):
            if map[j][i] == "S":
                start = (j, i)
                map[j][i] = "."
            elif map[j][i] == "E":
                end = (j, i)
                map[j][i] = "."

    pq = [(0, start[0], start[1], (0,1))]
    heapq.heapify(pq)

    while pq:
        (n, j, i, head) = heapq.heappop(pq)
        if (j, i) == end:
            print(n)
            break
        for dir in dirs:
            nj = j + dir[0]
            ni = i + dir[1]
            if map[nj][ni] == "." and map_cost[nj][ni] > (n+1+(1000 if dir!=head else 0)):
                map_cost[nj][ni] = (n+1+(1000 if dir!=head else 0))
                heapq.heappush(pq, ((n+1+(1000 if dir!=head else 0)), nj, ni, dir))
        for row in map_cost:
            print(" ".join((f"{elem:5}" if elem !=10_000 else "     ") for elem in row ))
        print(pq)
```
